## Vuelos

```python
flights = pd.read_parquet("../resources/flights.parquet")
flights.head()
```

Escoger los datos de enero 2024

```python
flights["STD"] = pd.to_datetime(flights["STD"])
flights["STA"] = pd.to_datetime(flights["STA"])
```

```python
flights = flights[flights["STD"].dt.year == 2024]
flights = flights[flights["STD"].dt.month == 1]
```

```python
# Registros nulos para cada campo
flights.isna().sum()
```

```python
# Borrar los datos que no tienen Aeronave
flights = flights.dropna(subset=["Aeronave"])
```

```python

flights_.isna().sum()
```

```python
# Ordenar por fecha
flights = flights.sort_values(by="STD")
```

Se crea una columna de `Semana` y `Mes` para an√°lisis posteriores.

```python

flights["Mes"] = flights["STD"].dt.month
flights["Semana"] = flights["STD"].dt.isocalendar().week
flights["DiaSemana"] = flights["STD"].dt.weekday + 1
flights["Hora"] = flights["STD"].dt.hour
flights
```

```python
flights["HourDuration"] = (flights["STA"] - flights["STD"]).dt.total_seconds() / 3600
flights.head()
```

```python
flights["HourDuration"].describe()
```

```python
(flights["HourDuration"] > 5).sum()
```

Se incluyen vuelos de itinerario sumados como hasta 27+ horas de viaje. Esto parece un error en los datos, por lo que se eliminan estos registros los cuales √∫nicamente representan el 1500 de alrededor de 121,724 de los datos.

```python
flights = flights[flights["HourDuration"] <= 5]
flights
```

```python
flights.to_parquet("../resources/flightstest_clean.parquet", index=False)
```

```python
# Se eliminan valores nulos pero
# realmente no hay
sales.dropna(inplace=True)
```

Se eliminan aquellos tipos de producto que no son de inter√©s para el an√°lisis pues no son tangibles

```python
excluded_types = [
    "Specials",
    "OFERTAS ",
    "Vivabus",
    "Transportaciones CUN",
    "Hertz.",
    "COMBOS CREW",
    "VIVA Taxis",
    "Transportaciones MTY",
    "VivaTransfer",
    "Antros",
    "VIVA PLAY",
    "Transportaciones TLC",
    "Alimentos Charter",
]

sales = sales[~sales.ProductType.isin(excluded_types)]
sales["ProductType"].value_counts()
```

```python
sales.info()
```

Se eliminan aquellas ventas que sean negaivas pues implican descuentos

```python
sales = sales[sales.TotalSales > 0]
```

```python
sales.drop(columns=["TotalSales"], inplace=True)
```

```python
sales.info()
```

```python
sales.reset_index(inplace=True)
```
